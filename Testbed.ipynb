```python
import numpy as np
import matplotlib.pyplot as plt


img = plt.imread("SupportDocs/court.jpg")
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(13, 6),gridspec_kw={'width_ratios': [4, 1]})
history = [(0,50,50),(-1,45,47),(1,38,37)]
ax1.imshow(img,extent=[-3, 3, 0, 7])
ax1.invert_yaxis()
ax1.set_aspect(0.5)
ax1.get_yaxis().set_visible(False)
ax1.plot(list(list(zip(*history))[0]),np.linspace(1,6,3),  linewidth=3, marker='o',markersize=12,color='yellow')
ax2.bar(['Self Score', 'Opponent Score'],[history[-1][1],history[-1][2]])
ax1.set_title('Game Progress',fontweight='bold',fontsize = 20)
ax2.set_title('Scores',fontweight='bold',fontsize = 20)
```
